```python
#1 라이브러리 임포트
import requests
from bs4 import BeautifulSoup
#2 웹페이지 지정
html = """
<html> 
  <body>
    <h1 id='title'>[1]크롤링이란?</h1>
    <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p>
    <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
  </body>
</html>
"""
#3 웹페이지 파싱
soup = BeautifulSoup(html, "html.parser")
#4 태그로 검색하여 필요한 html 코드 추출
data = soup.find('h1')
print(data)

#5 html로부터 필요한 데이터 추출
print(data.string)
print(data.get_text())
```

```python
#6 동일한 태그가 여러개인 경우
data1 = soup.find('p', class_='cssstyle')
data2 = soup.find('p', 'cssstyle')
data3 = soup.find('p', attrs={'align':'center'})
data4 = soup.find(id='body')

print(data1.string)
print(data2.string)
print(data3.string)
print(data4.string)
```

```python
#7 해당 태그의 모든 문장 가져오기
paragraph_data = soup.find_all('p')
for paragraph in paragraph_data:
  print(paragraph.get_text())
```
