#3. open API 크롤링 결과를 엑셀 보고서로 만들기

```python
import json
import requests
```

```python
cilent_id = X-Naver-Client-Id 키 값
client_secret = X-Naver-Client-Secret 키 값
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
import openpyxl

filepath = "/content/gdrive/My Drive/폴더명/news_samsung.xlxs"
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['번호', '제목', '링크'])
```

```python
#한번에 100개의 검색 결과만 가져올 수 있으므로 for문을 통해 더 많은 검색 결과를 가져온다.

start_num, num = 1, 0

for index in range(10):
  number = start_num + (index*100)
  naver_open_api = "https://openapi.naver.com/v1/search/news.json?query=samsung&display=100&start="+str(number)
  #print(naver_open_api)

  header_params = {"X-Naver-Client-Id":cilent_id, "X-Naver-Client-Secret" : client_secret}
  res = requests.get(naver_open_api, headers=header_params)

  if res.status_code == 200:
    data = res.json()
    for item in data['items']:
      num += 1
      #print(num, item['title'], item['description'])
      excel_sheet.append([num, item['title'], item['description']])
  else:
    print("Error Code :", res.status_code)

excel_file.save(filepath)
excel_file.close()
```
