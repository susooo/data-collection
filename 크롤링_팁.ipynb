###**2) 더 큰 범위의 태그를 이용**

```python
#1 라이브러리 임포트
import requests
from bs4 import BeautifulSoup
#2 웹페이지 지정
res = requests.get('https://www.w3schools.com/css/')
#3 웹페이지 파싱
soup = BeautifulSoup(res.content, "html.parser")
#4 태그로 검색하여 필요한 데이터 추출
data = soup.find_all('a', "w3-bar-item w3-button")
for item in data:
  print(item.get_text())
```

① find()로 더 크게 감싸는 상위 HTML 태그로 추출

```python
data1 = soup.find('div', "w3-col l3 m6")
```

② 추출된 데이터에서 find_all()로 다시 원하는 부분을 추출

```python
data2 = data1.find_all('a', "w3-bar-item w3-button")
for item in data2:
  print(item.get_text())
```

###**3) 데이터 가공하기**

① 파이썬 strip() 함수 이용하기

```python
data2 = data1.find_all('a', "w3-bar-item w3-button")
for item in data2:
  print(item.get_text().strip())
```

② 파이썬 split() 함수 이용하기

```python
data2 = data1.find_all('a', "w3-bar-item w3-button")
for item in data2:
  print(item.get_text().split())
```

```python
for item in data2:
  print(item.get_text().split()[1])
```

```python
#번호와 함께 출력
for index, item in enumerate(data2):
  print(str(index)+'.', item.get_text().split()[1])
```

###**6) 크롤링한 데이터 다시 크롤링하기**

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("http://corners.gmarket.co.kr/Bestsellers")
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
#크롤링 하다보면 None 값 나오거나 원하는 값과 다른 값이 나올 수 있다.  
#상위 태그로 select하고 변수로 저장하고 여러번 select하면 원하는 결과를 정확하게 얻을 수 있을 것이다.
data = soup.select('div.best-list')
bestitems = data[0]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
  title = product.select_one('a.itemname')
  price = product.select_one('div.item_price strong span')
  print(str(index+1)+'.', title.get_text(), '/ 할인가 : ', price.get_text(),end='')

  res_info = requests.get(title['href'])
  soup_info = BeautifulSoup(res_info.content, 'html.parser')
  company_name = soup_info.select_one( 'div.item-topinfo_headline > p > span.text__seller > a')
  print(' / 판매 회사명 : ', company_name.get_text())
```

```python
#더 나아가기 -> 엑셀파일에 저장하는 코드
import requests, openpyxl
from bs4 import BeautifulSoup

filepath = "/content/gdrive/My Drive/폴더명/news_samsung.xlxs"
excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['랭킹', '상품명', '판매 가격', '판매 회사명'])

res = requests.get("http://corners.gmarket.co.kr/Bestsellers")
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.select('div.best-list')
bestitems = data[0]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
  title = product.select_one('a.itemname')
  price = product.select_one('div.item_price strong span')

  res_info = requests.get(title['href'])
  soup_info = BeautifulSoup(res_info.content, 'html.parser')
  company_name = soup_info.select_one( 'div.item-topinfo_headline > p > span.text__seller > a')
  excel_sheet.append([str(index+1)+'.', title.get_text(), price.get_text(), company_name.get_text()])

excel_file.save('Bestproduct_Gmarket.xlsx')
excel_file.close()
```
